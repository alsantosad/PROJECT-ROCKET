####***ATIVIDADE***
> Cria√ß√£o do Dataframe, Calculo da M√©dia, Cria√ß√£o de uma nova tabela e exibi√ß√£o do Resultado.

Foi solicitada a cria√ß√£o de um DataFrame chamado nota com notas de tr√™s alunos (Ana, Ant√¥nio, Beatriz) em quatro disciplinas (Matem√°tica, F√≠sica, Qu√≠mica, Biologia).
C√°lculo da M√©dia por Aluno:

Foi solicitado o c√°lculo da m√©dia das notas para cada aluno e a adi√ß√£o de uma coluna chamada 'M√©dia' ao DataFrame.
C√°lculo da M√©dia Geral da Turma:

Foi solicitado o c√°lculo da m√©dia geral de todas as notas na tabela, incluindo as m√©dias individuais dos alunos.
Exibi√ß√£o dos Resultados:

Foi solicitada a exibi√ß√£o da tabela com as m√©dias das notas de cada aluno e a m√©dia geral da turma.

````python
# Importando as Bibliotecas
import pandas as pd
import numpy as np

#Criando uma estrutura de dados
notas_data = {'Matem√°tica': [6, 5, 5],
              'F√≠sica': [4, 7, 6],
              'Qu√≠mica': [5, 5, 7],
              'Biologia': [8, 5, 4]}

#Criando um DataFrame das notas
nota = pd.DataFrame(notas_data, index=['Ana', 'Ant√¥nio', 'Beatriz'])

# Calcular a m√©dia ao longo das linhas (axis=1)
nota['M√©dia'] = nota.mean(axis=1)

# Calcular a m√©dia geral de todas as notas na tabela
media_geral_turma = nota.values.mean()

# Exibir a nova tabela com as m√©dias das notas, a m√©dia geral da turma e a m√©dia geral de todas as notas
print(nota)
print("\nM√©dia Geral da Turma:", media_geral_turma)
````


A express√£o walmart.loc[walmart['Holiday_Flag'] == 1] √© usada para selecionar todas as linhas no DataFrame walmart onde a coluna 'Holiday_Flag' √© igual a 1. Essa abordagem √© √∫til quando voc√™ deseja filtrar as informa√ß√µes com base em uma condi√ß√£o espec√≠fica, neste caso, quando o valor da coluna 'Holiday_Flag' √© 1.

````python
walmart.loc[walmart['Holiday_Flag'] == 1]
````

A express√£o walmart.loc[walmart['Temperature'] > 90] √© utilizada para selecionar todas as linhas no DataFrame walmart onde os valores na coluna 'Temperature' s√£o maiores que 90. Isso cria um novo DataFrame que inclui apenas as entradas que atendem a essa condi√ß√£o espec√≠fica. Essa abordagem √© √∫til quando voc√™ deseja filtrar informa√ß√µes com base em crit√©rios espec√≠ficos, neste caso, quando a temperatura √© superior a 90.

````python
walmart.loc[(walmart['Temperature'] > 90)]
````

A express√£o walmart.loc[(walmart['Temperature'] > 90) & (walmart['Holiday_Flag'] == 1)] √© usada para selecionar as linhas no DataFrame walmart que atendem a duas condi√ß√µes simultaneamente:

A temperatura (na coluna 'Temperature') √© maior que 90.
O sinalizador de feriado (na coluna 'Holiday_Flag') √© igual a 1.
O operador & representa a condi√ß√£o l√≥gica "E" entre as duas condi√ß√µes. O resultado √© um novo DataFrame contendo apenas as entradas que satisfazem ambas as condi√ß√µes. Essa abordagem √© √∫til quando voc√™ deseja filtrar dados com base em m√∫ltiplos crit√©rios simultaneamente.

````python
walmart.loc[(walmart['Temperature'] > 90) & (walmart['Holiday_Flag'] == 1)]
````

A express√£o walmart.query('Holiday_Flag == 1') usa o m√©todo query do pandas para selecionar todas as linhas no DataFrame walmart onde o valor na coluna 'Holiday_Flag' √© igual a 1. Essa abordagem oferece uma maneira mais concisa de realizar essa filtragem, sem a necessidade de usar o m√©todo loc e os operadores l√≥gicos diretamente.

Essencialmente, a fun√ß√£o query permite que voc√™ especifique uma express√£o condicional de maneira mais leg√≠vel e compacta. O resultado √© um novo DataFrame contendo apenas as entradas que atendem √† condi√ß√£o especificada.

````python
walmart.query('Holiday_Flag == 1')
````

A express√£o walmart.query('39 <= Temperature > 50') procura no DataFrame walmart as linhas onde o valor na coluna 'Temperature' est√° entre 39 (inclusive) e 50 (exclusive). A corre√ß√£o sugerida para evitar compara√ß√µes encadeadas √© usar operadores l√≥gicos separados.

````python
walmart.query('39 <= Temperature > 50')
````

A express√£o walmart.query('Store in [1, 3]') est√° correta. Ela utiliza o m√©todo query do pandas para selecionar as linhas no DataFrame walmart onde o valor na coluna 'Store' est√° presente na lista [1, 3]. Isso resultar√° em um novo DataFrame contendo apenas as entradas correspondentes √†s lojas 1 e 3.

````python
walmart.query('Store in [1,3]')
````

A express√£o walmart.query('Temperature > @tempe') utiliza o m√©todo query do pandas e a nota√ß√£o @ para referenciar vari√°veis externas. Neste caso, a vari√°vel tempe com valor 50.0 √© usada na condi√ß√£o para selecionar as linhas no DataFrame walmart onde o valor na coluna 'Temperature' √© maior que 50.0. Isso resulta em um novo DataFrame contendo apenas as entradas que atendem a essa condi√ß√£o espec√≠fica de temperatura.

````python
tempe = 50.0
walmart.query('Temperature > @tempe')
````

A express√£o walmart2 = walmart.copy() cria uma c√≥pia independente do DataFrame original walmart e atribui essa c√≥pia √† vari√°vel walmart2. 
A express√£o walmart2.drop(0) remove a linha com o √≠ndice 0 do DataFrame walmart2. Se desejar alterar o DataFrame original, voc√™ pode usar inplace=True. Caso contr√°rio, pode atribuir o resultado a uma nova vari√°vel para manter uma c√≥pia modificada.


A express√£o walmart2.drop(['Holiday_Flag','Unemployment'], inplace=True, axis='columns') remove as colunas 'Holiday_Flag' e 'Unemployment' do DataFrame walmart2.
