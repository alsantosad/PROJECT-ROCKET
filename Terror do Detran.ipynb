### * ***Importando e visualizando o arquivo*** *

 * Esta célula faz o upload do arquivo e exibe as primeiras linhas do DataFrame para verificarmos como os dados estão estruturados.

```python
from google.colab import files
import pandas as pd
import io

def carregar_arquivo():
    # Fazendo o upload do arquivo
    uploaded = files.upload()

    # Obtendo o nome do arquivo
    file_name = next(iter(uploaded))

    # Verificando a extensão do arquivo
    extensao = file_name.split('.')[-1]

    # Lidando com diferentes extensões
    if extensao.lower() == 'xlsx':
        # Para arquivos Excel (.xlsx)
        dados = pd.read_excel(io.BytesIO(uploaded[file_name]), engine='openpyxl')
    elif extensao.lower() == 'csv':
        # Para arquivos CSV (.csv)
        dados = pd.read_csv(io.StringIO(uploaded[file_name].decode('utf-8')))
    elif extensao.lower() in ['txt', 'text']:
        # Para arquivos de texto (.txt)
        dados = pd.read_csv(io.StringIO(uploaded[file_name].decode('utf-8')), delimiter='\t')  # Altere o delimitador conforme necessário
    else:
        raise ValueError(f"Extensão de arquivo não suportada: {extensao}")

    return dados

# Carregando o arquivo
dados = carregar_arquivo()
```

### ***Corrigindo a coluna venda***

 * Aqui, corrigimos os valores na coluna "Valor de Venda" para garantir que estejam no formato float.

```python
# Corrigindo valores em "Valor de Venda"
dados['Valor de Venda'] = dados['Valor de Venda'].apply(lambda x: float(str(x).replace(',', '.')) if pd.notna(x) else x)

# Visualizando as primeiras linhas após a correção
dados.head()
```

### * ***Exclusão da coluna "Valor de Compra"*** *

 * A coluna foi excluída porque não haviam dados contidos nela, o que não agrega para a análise dos dados.

```python
# Excluindo a coluna "Valor de Compra"
dados = dados.drop(columns=['Valor de Compra'], errors='ignore')

# Visualizando as primeiras linhas após a exclusão
dados.head()
```

### * ***Conversão da coluna "Data da Coleta"*** *

 * Aqui, convertemos a coluna "Data da Coleta" para o formato de data.

```python
# Convertendo a coluna "Data da Coleta" para o formato de data
dados['Data da Coleta'] = pd.to_datetime(dados['Data da Coleta'], errors='coerce')

# Visualizando as primeiras linhas após a conversão
dados.head()
```

### * ***Eliminação de duplicatas*** *

 * Duplicatas são removidas do DataFrame. Isso é feito para que os dados, ao final da análise, possam ser bem apreciados.

```python
# Eliminando duplicatas
dados = dados.drop_duplicates()

# Visualizando as primeiras linhas após a remoção de duplicatas
dados.head()
```

### * ***Eliminação de linhas com dados faltantes*** *

 * Linhas com dados faltantes nas colunas especificadas são removidas. Isso não se aplica à coluna "Complemento", visto que o preenchimento deste dado é opcional. A coluna de Venda não apresentou dados faltantes, então não foi necessário fazer substituições por médias e medianas.

```python
# Eliminando linhas com dados faltantes
dados = dados.dropna(subset=['Regiao - Sigla', 'Estado - Sigla', 'Municipio', 'Revenda', 'CNPJ da Revenda', 'Nome da Rua', 'Numero Rua', 'Bairro', 'Cep', 'Produto', 'Data da Coleta', 'Valor de Venda', 'Unidade de Medida', 'Bandeira'])

# Visualizando as primeiras linhas após a remoção de linhas com dados faltantes
dados.head()
```

### * ***Eliminação de linhas com dados faltantes na coluna "Numero Rua"*** *

 * Linhas com dados faltantes na coluna "Numero Rua" são removidas. É de compreensão de todos que a númeração é algo importante. Como a base de dados possui mais de 100 mil amostras, cerca de 258 linhas sem númeração é passivo de limpeza.

```python
# Eliminando linhas com dados faltantes na coluna "Numero Rua"
dados = dados.dropna(subset=['Numero Rua'])

# Visualizando as primeiras linhas após a remoção de linhas com dados faltantes em "Numero Rua"
dados.head()
```
